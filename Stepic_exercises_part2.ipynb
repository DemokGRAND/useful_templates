```python
p = taxi.groupby('borough', as_index = True). \
     agg({'pickups':"sum"}) 
p.head()
```

```python
#Можем обратиться по индексу таким образом
p.pickups.EWR
```

```python
#В данном случае индексы это название района
taxi.groupby('borough', as_index = True). \
     agg({'pickups':"sum"}) \
     .pickups
#применим метод idxmin() который возвращает индекс меньшего числа в таблице
min_pickups = taxi.groupby('borough', as_index = True). \
     agg({'pickups':"sum"}) \
     .pickups.idxmin()
```

```python
min_pickups
```

```python
#https://stepik.org/lesson/375664/step/10?unit=363414
taxi.groupby(['borough','hday'], as_index = False) \
    .agg({'pickups':"mean"})
```

```python
taxi.head()
```

```python
#https://stepik.org/lesson/375664/step/11?unit=363414
taxi.groupby(['borough','pickup_month'], as_index = False) \
    .agg({'pickups':"sum"}) \
    .sort_values('pickups', ascending = False)
```

```python
def temp_to_celcius(in_df):
    out_df = ((in_df - 32) * 5)/9
    return out_df
    
```

```python
taxi.temp
```

```python
s = temp_to_celcius(taxi.temp)
```
